```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_wine
from sklearn.model_selection import cross_val_score


data = load_wine(as_frame=True)

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size= 0.2, random_state = 42)

metricList = ['minkowski', 'manhattan']
weightsList = ['uniform', 'distance']
kList = []
for i in range(1,50): #홀수만
    if(i%2 == 1):
        kList.append(i)
```

## No Cross Validation

```python
scaled_noncv_acc = pd.DataFrame(columns=['weights', 'metrics', 'k', 'train_acc', 'test_acc'])

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

for w in weightsList:
    for m in metricList:
        for k in kList:
            knn_clf = KNeighborsClassifier(weights = w, n_neighbors= k, metric = m)
            knn_clf.fit(X_train_scaled,y_train)
            y_train_pred = knn_clf.predict(X_train_scaled)
            y_test_pred = knn_clf.predict(X_test_scaled)
            scaled_noncv_acc = scaled_noncv_acc.append(pd.DataFrame([[w, m, k, accuracy_score(y_train, y_train_pred),accuracy_score(y_test, y_test_pred)]],
                                              columns=['weights', 'metrics', 'k', 'train_acc', 'test_acc']))           

high_train_acc = scaled_noncv_acc[scaled_noncv_acc['train_acc'] == max(scaled_noncv_acc['train_acc'])] ; high_train_acc
```

```python
high_train_acc['test_acc'].mean()
```

## Cross Validation

```python
scaled_cv_acc = pd.DataFrame(columns=['weights', 'metrics', 'k', 'score'])

for w in weightsList:
    for m in metricList:
        for k in kList:
            knn_clf = KNeighborsClassifier(weights = w, n_neighbors= k, metric = m)
            score = cross_val_score(knn_clf, X_train, y_train, cv=5)
            # knn_clf.fit(X_train_scaled,y_train)
            # y_train_pred = knn_clf.predict(X_train_scaled)
            # y_test_pred = knn_clf.predict(X_test_scaled)
            scaled_cv_acc = scaled_cv_acc.append(pd.DataFrame([[w, m, k, score.mean()]],
                                              columns=['weights', 'metrics', 'k', 'score']))           

scaled_cv_acc[scaled_cv_acc['score'] == max(scaled_cv_acc['score'])]
```

# 2. 모델 학습

```python
X = boston.data
y = boston.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

## 2.1. 회귀계수 및 MSE, RMSE

```python
print("----------------------------")
print(f"절편: {lin_reg.intercept_}\n")
print(f"계수: {lin_reg.coef_}")
coeff = pd.Series(data=np.round(lin_reg.coef_, 1), index=boston.feature_names)
print(f"{coeff.sort_values(ascending=False)}")
print("----------------------------")
y_pred = lin_reg.predict(X_test)
print(f"Train-R2: {lin_reg.score(X_train, y_train)}")
print(f"Test-R2: {lin_reg.score(X_test, y_test)}")
print(f"MSE: {mean_squared_error(y_pred, y_test)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_pred, y_test))}")
print("----------------------------")
```

## 3.2. Log Transform

> **선형회귀에 성능을 높이기 위한 조건**
> 1. Linearity
> 2. No noise
> 3. No collinearity
> 4. Normal distribution
> 5. Scale

> *참고 [Top 3 Methods for Handling Skewed Data](https://towardsdatascience.com/top-3-methods-for-handling-skewed-data-1334e0debf45)*

* 위 시각화 자료(목차 1.0 데이터 시각화)를 보면 강한 skewness를 보이는 열들이 많다.
* 이를 제거하기위한 방법 중 대표적인 로그변환을 사용해 보았다.

```python
X = df.iloc[:,:13]
y = df.iloc[:,13:]

y =  np.log1p(y)
for col in X.columns:
    if np.abs(X[col].skew()) > 0.3:
        X[col] = np.log1p(X[col])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

LT_reg = LinearRegression().fit(X_train_scaled, y_train)

print("----------------------------")
print(f"절편: {LT_reg.intercept_}\n")
print(f"계수: {LT_reg.coef_}")
print("----------------------------")
y_pred = LT_reg.predict(X_test_scaled)
print(f"Train-R2: {LT_reg.score(X_train_scaled, y_train)}")
print(f"Test-R2: {LT_reg.score(X_test_scaled, y_test)}")
print(f"MSE: {mean_squared_error(y_pred, y_test)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_pred, y_test))}")
print("----------------------------")
```

* MSE이 값이 극단적으로 낮아지긴 하였으나, 이는 모델의 성능개선 보다는 로그변환 자체로 수치가 작아진 것이 원인으로 보인다.
* 그러나 훈련셋과 테스트셋의 결정계수값이 기존 모델 대비 6~7%p 상승하였으므로 성능이 개선되었다고 판단된다.

## 3.3. Log Transform +
* 데이터를 전처리 하는것에서 가장 유의미한 성능향상을 보여 추가적인 데이터전처리를 해주었다.
* 범주형 데이터 'CHAS'열을 제거하였다. (목차 1.0 데이터 시각화 참고)
* 타겟값 'y'에서 이상치를 제거해 주었다.

```python
df2 = df[~(df['y'] >= 50.0)]
print(np.shape(df2))

X = df2.iloc[:,:13]
X = X[['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
y = df2.iloc[:,13:]

y =  np.log1p(y)
for col in X.columns:
    if np.abs(X[col].skew()) > 0.3:
        X[col] = np.log1p(X[col])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

LT_regP = LinearRegression().fit(X_train_scaled, y_train)

print("----------------------------")
print(f"절편: {LT_regP.intercept_}\n")
print(f"계수: {LT_regP.coef_}")
print("----------------------------")
y_pred = LT_regP.predict(X_test_scaled)
print(f"Train-R2: {LT_regP.score(X_train_scaled, y_train)}")
print(f"Test-R2: {LT_regP.score(X_test_scaled, y_test)}")
print(f"MSE: {mean_squared_error(y_pred, y_test)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_pred, y_test))}")
print("----------------------------")
```

* 훈련셋과 테스트셋의 결정계수값이 기존 모델 대비 5~8%p 상승하였다.

* MSE, RMSE값의 변화는 성능개선으로 인한 효과보다는 로그변환으로 수치자체가 낮아져 나타난 변화로 보인다.
* 그럼에도 로그변환과 데이터 전처리를 추가로 해준 **'Log Transform+'** 모델에서 가장 높은 결정계수값 상승을 보인다.
